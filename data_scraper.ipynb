```python
import requests
import json
```

```python
#Endpoint for getting general data about the election first param is id, and second is language

general_data_api = "https://www.izbori.ba/api_2018/administration_electionresultcontroller_getgeneraldata/{}/3"
```

```python
valid_data = []
for i in range(40):
    data = requests.get(general_data_api.format(i)).json()
    if data['nameElection']:
        data['id'] = i
        valid_data.append(data)
```

```python
race_api = "https://www.izbori.ba/api_2018/administration_racecontroller_getbyelectionresultId/{}/1"
codes_dict = {}
for data in valid_data:
    races = requests.get(race_api.format(data['id'])).json()
    data['codes'] = []
    for d in races:
        data['codes'].append(int(d['code']))
        if not d['code'] in codes_dict.keys():
            codes_dict[d['code']] = d['name']
```

```python
database_name_api = "https://www.izbori.ba/api_2018/administration_electionresultcontroller_getdatabasename/{}"
```

```python
for data in valid_data:
    database = requests.get(database_name_api.format(data['id'])).json()
    data['database'] = str(database)
```
